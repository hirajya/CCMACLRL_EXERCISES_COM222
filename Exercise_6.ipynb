## Data Preprocessing

```python


# Define a function to remove outliers using the IQR method and report counts
def remove_outliers_iqr(df, column):
    # Count rows before removing outliers
    before_count = df.shape[0]

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the DataFrame to remove outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Count rows after removing outliers
    after_count = df_filtered.shape[0]

    # Print the counts
    print(f"Removing outliers from '{column}':")
    print(f"Before: {before_count}, After: {after_count}\n")

    return df_filtered

# Remove outliers for the 'Age' and 'Fare' columns
df = remove_outliers_iqr(df, 'Age')
df = remove_outliers_iqr(df, 'Fare')
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin', 'Embarked'],inplace=True)
dt.drop(columns=['PassengerId','Name','Ticket','Cabin', 'Embarked'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Convert 'Sex' to numerical values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
dt['Sex'] = dt['Sex'].map({'male': 0, 'female': 1})

df['Fare'].fillna(df['Fare'].median(), inplace=True)
dt['Fare'].fillna(dt['Fare'].median(), inplace=True)


# Handle missing values in 'Age' (you can choose a different strategy if needed)
df['Age'].fillna(df['Age'].median(), inplace=True)
dt['Age'].fillna(dt['Age'].median(), inplace=True)
```

```python
def convert_age(df, cut_off_values, label_names):
    df["Age"] = df["Age"].fillna(-0.5)
    df["Age_Categories"] = pd.cut(df["Age"], cut_off_values, label_names)
    return df

cut_off_values = [-1, 0, 5, 12, 18, 35, 60, 100]
label_names = ["Missing", "Infant", "Child", "Teenager", "Young Adult", "Adult", "Senior"]

df = convert_age(df, cut_off_values, label_names)
dt = convert_age(dt, cut_off_values, label_names)

# age_plot = df.pivot_table(index = "Age_Categories", values = "Survived")
# age_plot.plot.bar()
# plt.show()
```

```python
def create_dummies(df, column_name):
    dummies = pd.get_dummies(df[column_name], prefix = column_name)
    df = pd.concat([df, dummies], axis = 1)
    return df

df = create_dummies(df, "Pclass")
dt = create_dummies(dt, "Pclass")

df = create_dummies(df, "Age_Categories")
dt = create_dummies(dt, "Age_Categories")

df = create_dummies(df, "Sex")
dt = create_dummies(dt, "Sex")

df.head(10)
dt.head(10)
```
