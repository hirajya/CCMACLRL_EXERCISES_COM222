5. Calculate correlation coefficient and p-value between Miles Traveled and Travel Time


```python
# Calcule correlation coefficient and p-value between 'Miles Traveled' and 'Travel Time'
result = pearsonr(df['Miles Traveled'], df['Travel Time'])
print(f"Miles Traveled and Travel Time")
print(f"Correlation coefficient: {result[0]}")
print(f"P-value: {result[1]}")
```

### Building Different Regression Models


#### Model 1: Miles Traveled and Travel Time


16. Set independent and dependent variables

```python
X = df[['Miles Traveled']]
X = sm.add_constant(X)
y = df['Travel Time']
```

17. Build and train the model


```python
mlr_model = LinearRegression()

mlr_model.fit(X, y)

y_pred = mlr_model.predict(X)
print(f"Intercept: {mlr_model.intercept_}")
print(f"Coefficient: {mlr_model.coef_[0]}")
```

18. Display the statistical summary


```python
olsmod = sm.OLS(y, X).fit()
olsmod.summary()   
```

19. Display the r-squared


```python
print(olsmod.rsquared)
```
